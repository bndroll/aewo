# Подготовка и обучение моделей

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
```

Функция для рисования матрицы ошибок(она перевернутая, лучше не вставлять)

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

Датафреймы для обучения классификатора

```python
exhauster1_class = exhauster1.copy()
exhauster2_class = exhauster2.copy()
exhauster3_class = exhauster3.copy()
exhauster4_class = exhauster4.copy()
exhauster5_class = exhauster5.copy()
exhauster6_class = exhauster6.copy()
```

Замена непрерывных значений переменной is_work на категориальные 1 и 0

```python
exhauster1_class['is_work'] = np.where(exhauster1_class['is_work'] > 0.2, 1, 0)
exhauster2_class['is_work'] = np.where(exhauster2_class['is_work'] > 0.26, 1, 0)
exhauster3_class['is_work'] = np.where(exhauster3_class['is_work'] > 1.05, 1, 0)
exhauster4_class['is_work'] = np.where(exhauster4_class['is_work'] > 1.3, 1, 0)
exhauster5_class['is_work'] = np.where(exhauster5_class['is_work'] > 0.75, 1, 0)
exhauster6_class['is_work'] = np.where(exhauster6_class['is_work'] > 0.9, 1, 0)
```

## Создание отдельной модели для каждого эксгаустера

### 1 модель

```python
X1 = exhauster1_class.drop(['is_work'], axis=1)
y1 = exhauster1_class['is_work']
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.33)
```

```python
clf1 = tree.DecisionTreeClassifier()
```

```python
clf1.fit(X1_train, y1_train)
```

```python
y1_pred = clf1.predict(X1_test)
```

```python
y1_pred = np.where(y1_pred < 0.5, 0, 1)
```

```python
precision_score(y1_test, y1_pred)
```

```python
recall_score(y1_test, y1_pred)
```

```python
f1_score(y1_test, y1_pred)
```

```python
font = {'size' : 15}

plt.rc('font', **font)

cnf_matrix = confusion_matrix(y1_test, y1_pred)
plt.figure(figsize=(10, 8))
plot_confusion_matrix(cnf_matrix, classes=['Not working', 'Working'],
                      title='Confusion matrix')
plt.show()
```
